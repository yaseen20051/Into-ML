Series

```python
from pandas import Series 

### Different Datatypes 

int_data = [100,102,104]
int_series = pd.Series(int_data)

float_data = [10.5,20.6,11.5]
float_series = pd.Series(float_data)

object_data = ["Y","a","s"]
object_series = pd.Series(object_data)

bool_data= [True,False]
bool_series = pd.Series(bool_data)

print(f"int datatype : \n{int_series}\n\nfloat datatype : \n{float_series}\n\nobject datatype\n{object_series}\n\nbool datatype\n\n{bool_series}")

### Changing Index  

data = [100,102,104]

changed_labels = Series(data,index = ["User 1","User 2","User 3"])
changed_labels

### loc property 

        ## locate using label
value_user = changed_labels.loc["User 1"]
print(f"\nValue of User 1 (label located)= ${value_user}")

        ## locate using index (staring from zero)
index_located_user_value = changed_labels.iloc[0]
print(f"Value of User 1 (index located)= ${index_located_user_value}")

        ## Changing label
last_value = value_user
changed_labels.loc["User 1"] = 9032
value_user = changed_labels.loc["User 1"]
print(f"\nlast value of user 1 id = {last_value}\ncurrent value after changing using the loc property = {value_user} \n")

        ## filter by value 
prices = [150.5,100,50,20,260,330,550,120]
series_prices = Series(prices,index=["item 1","item 2","item 3","item 4","item 5","item 6","item 7","item 8"])
sum = 0
avg = 0
for price in prices:
    sum = sum + price
avg = sum / len(prices)
print(series_prices[series_prices>avg],"\n")
        ## using dictonary
calories = {"Day 1": 1750, "Day 2":2100,"Day 3":1600} 
## keys will be the labels so no need to insert indeces or labels 
series_calories  = Series(calories)  
print(series_calories)

series_calories.loc["Day 2"] -= 200
print(series_calories.loc["Day 2"],"\n")
print(series_calories.iloc[2],"\n")

print("Cheated days")
print(series_calories[series_calories>1700])
```

DataFrames

```python
from pandas import DataFrame 

data = {
        "Name":["Spongebob","Patrick","Squidward"],
        "Age":[30,50,35]
    }
df = DataFrame(data)
print(df,"\n")

### same manipulation as the series 

    ## changing labels
df = DataFrame(data,index=["Employee 1","Employee 2","Employee 3"])
print(df,"\n")
    ## loc property by (locate by label =)
print(df.loc["Employee 2"],"\n")
    ## iloc property (locate by index)
print(df.iloc[2],"\n")
    ## filtering by vaule and Accessing a specific feature 
print(df[df["Age"]>30],"\n")

### Adding a new columns
df["Job"] = ["Cook","N/A","Cashier"]
print(df,"\n")
### Adding a new row 
        ## new row = pd.DataFrame([{"feature 1":value,"feature 2":value},{second row (second training example )}])
new_row = DataFrame([{"Name":"Sandy","Age":25,"Job":"Scientist"},{"Name":"Yaseen","Age":20,"Job":"ML Engineer"}],index=["Employee 4","Employee 5"])
# DONT FORGET TO CONCATENATE THE NEW RULE OR THE NEW EXAMPLES
df = pd.concat([df,new_row])
print(df,"\n")
    ## other way to add a row (unsuccessful) different examples must be added independently from eachother 
# new_row = DataFrame([{
#                         "Name":["Yahia","Salem"],
#                         "Age":[18,36],
#                         "Job":["Student","Manger"]
#                     }])
# df = pd.concat([df,new_row])
# print(df,"\n")
```
